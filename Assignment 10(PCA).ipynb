### Q3:  
#### We will apply principal component analysis and linear discriminant analysis on iris data.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.preprocessing import StandardScaler

iris = datasets.load_iris()
X = iris.data
y = iris.target
X_scale = StandardScaler().fit_transform(X)
```

```python
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis


pca = PCA(n_components=2)
lda = LinearDiscriminantAnalysis(n_components=2)

plt.figure(figsize=(16,8))

plt.subplot(1, 3, 1, aspect=1)
pca.fit(X_scale, y)
X_embedded = pca.transform(X_scale)
plt.scatter(X_embedded[:, 0], X_embedded[:, 1], c=y, cmap='Set1')
plt.title(f'PCA')

plt.subplot(1, 3, 2, aspect=1)
pca.fit(X_scale)
X_embedded = pca.transform(X_scale)
plt.scatter(X_embedded[:, 0], X_embedded[:, 1], c=y, cmap='Set1')
plt.title(f'PCA')

plt.subplot(1, 3, 3, aspect=1)
lda.fit(X_scale, y)
X_embedded = lda.transform(X_scale)
plt.scatter(X_embedded[:, 0], X_embedded[:, 1], c=y, cmap='Set1')
plt.title(f'LDA')
plt.show()
```

(a) What effect does the `y` do in the `pca.fit(X_scale, y)` in the block above?

> Ans:  沒有影響，因为 PCA 是非監督式學習

(b) Which of the method tries to identify combination of attributes that account for the most variance ***in the data***, PCA or LDA? Which of the method tries to identify combination of attributes that account for the most variance ***between classes***, PCA or LDA?

> Ans:  
the most variance in the data : PCA<br>
the most variance between classes : LDA<br>
